Isolation Forest

Isolation Forests (iForest) is an ensemble method that isolates anomalies instead of profiling 'normal' data points. It works by constructing random decision trees and using the path lengths required to isolate a point as the criterion to flag anomalies. Since anomalies tend to be different from the bulk of data, they appear in shorter paths.

```python
from sklearn.ensemble import IsolationForest
from sklearn.datasets import load_wine # Import the Wine dataset
import numpy as np

# Load a dataset from scikit-learn
wine = load_wine()
X = wine.data # Use the features of the Wine dataset

# Initialize and train the Isolation Forest model
# contamination='auto' is reasonable, or you could set a fraction (e.g., 0.01)
clf = IsolationForest(contamination='auto', random_state=42) # Added random_state for reproducibility
clf.fit(X)

# Predict anomalies (-1 for outlier, 1 for inlier)
anomalies = clf.predict(X)

# Print the number of predicted anomalies
print(f"Total number of data points: {len(X)}")
print(f"Number of predicted anomalies (-1): {np.sum(anomalies == -1)}")
print(f"Number of predicted inliers (1): {np.sum(anomalies == 1)}")
```
